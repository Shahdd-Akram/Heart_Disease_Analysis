```python
# a function to remove outliers
def remove_outliers_iqr(df):
    """Removes outliers from all numerical columns using the IQR method."""
    df_cleaned = df.copy() 
    
    for column in df.select_dtypes(include=['number']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    
    return df_cleaned  


df_cleaned = remove_outliers_iqr(df)
print(df_cleaned)
```

```python
# Mean, Median, and Standard Deviation of Cholesterol (chol)
chol_mean = df["chol"].mean()
chol_median = df["chol"].median()
chol_std = df["chol"].std()

print(f"Cholesterol - Mean: {chol_mean}, Median: {chol_median}, Std Dev: {chol_std}")

# Mean Blood Pressure (trestbps) for patients with and without heart disease
mean_bp_heart_disease = df[df["target"] == 1]["trestbps"].mean()
mean_bp_no_disease = df[df["target"] == 0]["trestbps"].mean()

print(f"Mean Blood Pressure (with heart disease): {mean_bp_heart_disease}")
print(f"Mean Blood Pressure (without heart disease): {mean_bp_no_disease}")

# Max and Min values for Maximum Heart Rate (thalach)
thalach_max = df["thalach"].max()
thalach_min = df["thalach"].min()

print(f"Max Thalach: {thalach_max}, Min Thalach: {thalach_min}")
```

```python
# Sort patients by cholesterol levels
df_sorted = df.sort_values(by="chol", ascending=True)
print("Sorted by Cholesterol:\n", df_sorted.head())

# Identify patients with cholesterol above 300 mg/dL
high_chol_patients = df[df["chol"] > 300]
print("Patients with Cholesterol > 300:\n", high_chol_patients)

# Find patients older than 60 with abnormal ECG (restecg > 0)
abnormal_ecg_patients = df[(df["age"] > 60) & (df["restecg"] > 0)]
print("Patients older than 60 with abnormal ECG:\n", abnormal_ecg_patients)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

def visualize_heart_disease(df):
    """
    this function takes the df and plots the following visualizations:
    1. Histogram of cholesterol levels.
    2. Scatter plot of Age vs. Maximum Heart Rate.
    3. Bar chart comparing patients with and without heart disease.
    4. 3D Plot of Cholesterol, Age, and Heart Disease status.
    5. Pie chart showing the proportion of different Chest Pain Types.
    """
    plt.figure(figsize=(15, 12))
    
    # Histogram
    plt.subplot(2, 3, 1)
    sns.histplot(df["chol"], bins=20, kde=True, color='blue')
    plt.title("Cholesterol Distribution")
    plt.xlabel("Cholesterol (mg/dL)")
    plt.ylabel("Frequency")
    
    # Scatter Plot
    plt.subplot(2, 3, 2)
    sns.scatterplot(x=df["age"], y=df["thalach"], hue=df["target"], palette=['red', 'green'])
    plt.title("Age vs. Maximum Heart Rate")
    plt.xlabel("Age")
    plt.ylabel("Maximum Heart Rate")
    
    # Bar Chart
    plt.subplot(2, 3, 3)
    sns.countplot(x=df["target"], palette=['blue', 'orange'])
    plt.title("Heart Disease Presence")
    plt.xlabel("Target (0 = No, 1 = Yes)")
    plt.ylabel("Count")
    
    # 3D Plot
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(df["chol"], df["age"], df["target"], c=df["target"], cmap='coolwarm')
    ax.set_xlabel("Cholesterol")
    ax.set_ylabel("Age")
    ax.set_zlabel("Heart Disease (0/1)")
    ax.set_title("3D Plot: Cholesterol, Age & Heart Disease")
    
    # Pie Chart
    plt.subplot(2, 3, 4)
    cp_counts = df["cp"].value_counts()
    plt.pie(cp_counts, labels=cp_counts.index, autopct='%1.1f%%', colors=['lightblue', 'orange', 'green', 'red'])
    plt.title("Chest Pain Type Distribution")
    
    plt.tight_layout()
    plt.show()


visualize_heart_disease(df)
```
